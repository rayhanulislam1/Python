```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 10)
x = 5 * np.sin(2 * np.pi * 500 * t + np.pi/2)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
#plt.title('Plot of x(t) = 5 * sin(2π * 500t + 90)')
plt.grid(True)
plt.show()
```
